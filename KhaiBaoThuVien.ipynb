```python
# Kiểm tra dữ liệu thiếu ở mỗi cột
df.isnull().sum() # số dòng rỗng
```

```python
# Số dòng null của mỗi cột
null_counts = {
    "Rank": 0,
    "Name": 0,
    "Japanese_name": 10557,
    "Type": 0,
    "Episodes": 8994,
    "Studio": 6477,
    "Release_season": 14379,
    "Tags": 400,
    "Rating": 3131,
    "Release_year": 383,
    "End_year": 15641,
    "Description": 4,
    "Content_Warning": 16655,
    "Related_Mange": 10868,
    "Related_anime": 8432,
    "Voice_actors": 3186,
    "staff": 5490
}

# Tên của các cột
columns = list(null_counts.keys())

# Số lượng dòng
total_rows = 18495

# Số dòng null
null_values = list(null_counts.values())

# Tạo biểu đồ
plt.figure(figsize=(10, 6))
plt.barh(columns, null_values, color='skyblue')
plt.xlabel('Số dòng null')
plt.title('Số dòng null của mỗi cột')
plt.gca().invert_yaxis()  
plt.show()
```
